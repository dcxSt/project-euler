```python
import matplotlib.pyplot as plt
import pickle
import datetime as dt
import numpy as np
```

```python
t = open("times.pickle","rb")
times = pickle.load(t)
t.close()

start_time = times[0][1]
end_time = times[-1][1]
b_times=[]
for i in range(len(times[3:-1])):
    if i%2==0:
        b_times.append([ times[3:-1][i][0] , times[3:-1][i+1][1] - times[3:-1][i][1] ])
b_times.insert(0,[0,times[2][1]-times[1][1]])

[b,time_taken] = np.transpose(b_times)
b,time_taken = list(b),list(time_taken)
time_taken = [float(str(i)[5:]) for i in time_taken]
q = np.transpose([b,time_taken])
q = [[int(i[0]),i[1]] for i in q]
q.sort()
[b,time_taken] = np.transpose(q)
b = [int(i) for i in b]
```

```python
def isprime(n):
    for i in range(2,int(np.sqrt(n)+1)):
        if n%i==0:
            return False
    return True
```

```python
b_evens = [b[i] for i in range(0,len(b),2)]
log_b_evens = [np.log(i) for i in b_evens[:3400]]

b_odds = [b[i] for i in range(1,len(b),2)]
log_b_odds = [np.log(i) for i in b_odds[:3400]]

times_even = [time_taken[i] for i in range(0,len(time_taken),2)]
times_odd = [time_taken[i] for i in range(1,len(time_taken),2)]

b_primes_1000 = [i for i in b[1000:] if isprime(i)]
times_primes_1000 = [time_taken[i] for i in range(1001,len(time_taken),2) if isprime(b[i])]

b_primes = [i for i in b[2:] if isprime(i)]
log_b_primes = [np.log(i) for i in b_primes[:3400]]
times_primes = [time_taken[i] for i in range(2,len(time_taken)) if isprime(b[i])]
log_times_primes = [np.log(i) for i in b_primes[:3400]]

print(len(b_primes),
len(times_primes))
```

```python
plt.figure(figsize=(9,9))

#plt.subplot(221)
plt.title("the time it took for my script to calculate all the\n\
factors associated with a+ib for each complex 'row'")

plt.plot(b_evens[500:],times_even[500:],"x",label="b is even",color=(0,0,1,0.25))
plt.plot(b_odds[500:],times_odd[500:],"x",label="b is odd",color=(1,0,0,0.25))
plt.plot(b_primes_1000,times_primes_1000,"x",label="b is prime",color=(1,0,1,0.75))
plt.legend()
plt.xlabel("the complex componant b")
plt.ylabel("time taken in seconds for my algorithm to find sum of particular row")

plt.savefig("run-time.png")

"""
plt.subplot(222)
plt.title("the time it took for my script to calculate all the\n\
factors associated with a+ib for each complex 'row'")
"""
#plt.grid()

plt.show()
```
